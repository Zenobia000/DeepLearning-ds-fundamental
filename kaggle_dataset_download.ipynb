```python
import subprocess
import zipfile
import os

# 解壓縮並刪除ZIP文件
def extract_and_delete_zip(zip_file_path: str, extract_to_directory: str):
    """
    解压指定的zip文件到指定目录，然后删除该zip文件。

    参数:
    zip_file_path (str): zip文件的路径。
    extract_to_directory (str): 解压缩目标目录。

    返回:
    无返回值，但会在文件系统中产生效果（解压缩并删除zip文件）。
    """
    # 解压缩zip文件
    with zipfile.ZipFile(zip_file_path, "r") as zipref:
        zipref.extractall(extract_to_directory)
    
    # 删除原始zip文件
    os.remove(zip_file_path)
    print("下載的zip檔案已解壓縮並刪除。")



# 使用subprocess下载比赛数据
def download_competition_data(competition: str):
    zip_file_path = f"{competition}.zip"
    
    # 检查文件是否已存在
    if os.path.exists(zip_file_path):
        print(f"{zip_file_path} 已存在，跳過下载。")
        return
    
    command = f"kaggle competitions download -c {competition}"
    subprocess.run(command, shell=True)
    print("下载完成。")



if __name__ == "__main__":
    competition_name = 'dogs-vs-cats'
    # 假设当前工作目录是下载目录
    download_competition_data(competition_name)

    zip_file_path = f"{competition_name}.zip"
    extract_to_directory = "./data/cv/"

    # 解压并删除ZIP文件
    if os.path.exists(zip_file_path):
        extract_and_delete_zip(zip_file_path, extract_to_directory)
    else:
        print(f"未找到 {zip_file_path}，請確認資料集是否下載。")
```
